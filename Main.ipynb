```python
# Improved API request and data processing
import requests

# Define the API URL and headers

url = f"https://api.sportradar.com/tennis/trial/v3/en/competitions.json?api_key=bp7x57E6fYGPfcjgwxTVTK6CaFesccwdPX5bGXGE"
headers = {"accept": "application/json"}

# Function to fetch data from the API
def fetch_data(api_url, headers):
    try:
        response = requests.get(api_url, headers=headers)
        response.raise_for_status()  # Raise an error for bad status codes
        return response.json()
    except requests.exceptions.RequestException as e:
        print(f"Error fetching data from API: {e}")
        return None

# Fetch and process data
data = fetch_data(url, headers)
if data:
    try:
        competitions = data.get("competitions", [])
        competition_details = []  # List to store competition details
        category_details = []  # List to store category details

        for competition in competitions:
            # Extract competition details
            comp_detail = {
                "competition_id": competition.get("id"),
                "competition_name": competition.get("name"),
                "parent_id": competition.get("parent_id"),
                "type": competition.get("type"),
                "gender": competition.get("gender"),
                "category_id": competition.get("category", {}).get("id"),
            }
            competition_details.append(comp_detail)

            # Extract category details
            category = competition.get("category", {})
            category_detail = {
                "category_id": category.get("id"),
                "category_name": category.get("name"),
            }
            category_details.append(category_detail)

        print("Data processed successfully.")
    except Exception as e:
        print(f"Error processing data: {e}")
else:
    print("No data fetched from the API.")
```

Execute the following SQL queries:
1) Get all competitors with their rank and points.

2) Find competitors ranked in the top 5

3) List competitors with no rank movement (stable rank)

4) Get the total points of competitors from a specific country (e.g., Croatia)

5) Count the number of competitors per country

6) Find competitors with the highest points in the current week

```python
# 1) Get all competitors with their rank and points
cursor.execute("""
    SELECT c.competitor_id, c.name, cr.rank, cr.points
    FROM competitors c
    JOIN competitor_rankings cr ON c.competitor_id = cr.competitor_id
""")
all_competitors = cursor.fetchall()

print("All Competitors:", all_competitors)
```

```python
# 2) Find competitors ranked in the top 5
cursor.execute("""
    SELECT c.competitor_id, c.name, cr.rank, cr.points
    FROM competitors c
    JOIN competitor_rankings cr ON c.competitor_id = cr.competitor_id
    WHERE cr.rank <= 5
    ORDER BY cr.rank ASC
""")
top_5 = cursor.fetchall()

print("Top 5 Competitors:", top_5)
```

```python
# 3) List competitors with no rank movement
cursor.execute("""
    SELECT c.competitor_id, c.name, cr.rank, cr.movement
    FROM competitors c
    JOIN competitor_rankings cr ON c.competitor_id = cr.competitor_id
    WHERE cr.movement = 0
""")
stable_ranks = cursor.fetchall()

print("Stable Rank Competitors:", stable_ranks)
```

```python
# 4) Total points of competitors (example from Croatia)
cursor.execute("""
    SELECT c.country, SUM(cr.points) AS total_points
    FROM competitors c
    JOIN competitor_rankings cr ON c.competitor_id = cr.competitor_id
    WHERE c.country = 'Croatia'
    GROUP BY c.country
""")
points = cursor.fetchone()


print("Croatia Total Points:", points)
```

```python
# 5) Count the number of competitors per country
cursor.execute("""
    SELECT country, COUNT(*) AS total_competitors
    FROM competitors
    GROUP BY country
    ORDER BY total_competitors DESC
""")
competitors_per_country = cursor.fetchall()

print("Competitors Per Country:", competitors_per_country)
```

```python
# 6) Competitors with the highest points
cursor.execute("""
    SELECT c.competitor_id, c.name, cr.points
    FROM competitors c
    JOIN competitor_rankings cr ON c.competitor_id = cr.competitor_id
    WHERE cr.points = (SELECT MAX(points) FROM competitor_rankings)
""")
highest_points = cursor.fetchall()

print("Highest Points Competitors:", highest_points)
```
